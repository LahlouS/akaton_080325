### Dataprep

Since I want to be able to produce an overall similarity between a user and a product. user_description seems pretty important to me.
I will just drop all user that has no user_description. It will make the embedding step easier because I will not have to differentiate user that have description vs user that dont have description.
also 70% of user that dont have description also have 1 tweet in the total dataset so I presume they are not the user that carry the most information.
also, as predicted, user that dont have description carry way less informtion in the sens that they have an average like per tweet of 1 against 7 for donald and 10 for joe



Since I want to build a databased focus on describing a user, more than a tweet, I want to drop all the duplicate (user_id, tweet_id)

```python
donald_drop_null = donald[~(donald['user_description'].isnull())]
donald_drop_null = donald_drop_null[~(donald_drop_null['user_name'].isnull())]
donald_drop_null = donald_drop_null[~(donald_drop_null['user_name'].isin(["unknown", "N/A"]))]
print(len(donald_drop_null))
donald_drop_null = donald_drop_null.drop_duplicates(subset=['user_id', 'tweet_id'], keep='first')
donald_drop_null = donald_drop_null.reset_index(drop=True)
print(len(donald_drop_null))
sum.check_null_ish_values(donald_drop_null)
```

```python
joe_drop_null = joe[~(joe['user_description'].isnull())]
joe_drop_null = joe_drop_null[~(joe_drop_null['user_name'].isnull())]
joe_drop_null = joe_drop_null[~(joe_drop_null['user_name'].isin(["unknown", "N/A"]))]
joe_drop_null = joe_drop_null[~(joe_drop_null['user_description'].isin(["unknown", "N/A"]))]
print(len(joe_drop_null))
joe_drop_null = joe_drop_null.drop_duplicates(subset=['user_id', 'tweet_id'], keep='first')
joe_drop_null = joe_drop_null.reset_index(drop=True)

print(len(joe_drop_null))

sum.check_null_ish_values(joe_drop_null)
```

```python
import random

def generate_random_id():
	start_range = 10**17       # Minimum value for an 18-digit ID
	end_range = 2**63 - 1      # Maximum value for 64-bit signed integer

	# Generate a random integer within this range
	random_id = random.randint(start_range, end_range)
	return random_id

print(generate_random_id())
```
